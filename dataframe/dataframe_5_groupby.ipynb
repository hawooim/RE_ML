## df.groupby()

```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age', 'sex', 'class', 'fare', 'survived']]
df.head()
```

```python
for key, group in df.groupby('class'):
    print(key)
    print(group)
```

```python
df.groupby('class').mean()
```

```python
df.groupby('class').get_group('Third')
```

```python
df.loc[df['class'] == 'Third']
```

```python
for key, group in df.groupby(['class', 'sex']):
    print(key)
    print(group)
```

```python
df.groupby(['class', 'sex']).mean()
```

```python
df.groupby(['class', 'sex']).get_group(('Third', 'female'))
```

```python
df.loc[(df['class'] == 'Third') & (df['sex'] == 'female')]
```

## df.groupby().aggregation functions

```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age', 'sex', 'class', 'fare', 'survived']]
df.head()
```

```python
# aggregation
# mean(), max(), min(), sum(), count(), size(), var(), describe(), info(), first(), last()

df.groupby('class').std()
```

```python
df.groupby('class')['fare'].std()
```

## df.groupby().agg(aggregation function)

```python
def min_max(x):
    return x.max() - x.min()

df.groupby('class')[['age', 'fare']].agg(min_max)
```

## df.groupby().agg([func1, func2, ...])
## df.groupby().agg({col1: [func1, func2, ...], col2: [func3, ], ...})

```python
df.groupby('class').agg({'fare': ['min', 'max'], 'age': ['mean']})
```

## df.groupby().transform(function)

```python
# without using .transform

age_mean = df.groupby('class')['age'].mean()
age_mean
```

```python
age_std = df.groupby('class')['age'].std()
age_std
```

```python
for key, group in df.groupby('class')['age']:
    print(key)
    zscore = (group - age_mean.loc[key]) / age_std.loc[key]
    print(zscore)
    print()
```

```python
# using .transform

def z_score(x):
    return (x - x.mean()) / x.std()

age_zscore = df.groupby('class')['age'].transform(z_score)
age_zscore
```

## df.groupby().filter(conditional function)

```python
df.groupby('class').filter(lambda x: x.shape[0] >= 200)

# among three classes, it leaves only First and Third classes that satisfy the condition.
```

```python
df.groupby('class').filter(lambda x: x['age'].mean() < 30)

# It filters out the first class.
```

## df.groupby().apply(function)

```python
df.groupby('class').apply(lambda x: x.describe())
```

```python
def z_score(x):
    return (x - x.mean()) / x.std()

df.groupby('class')['age'].apply(z_score)
```

```python
age_filter = df.groupby(['class']).apply(lambda x: x['age'].mean() < 30)
age_filter
```

```python
for x in age_filter.index:
    if age_filter[x] == True:
        age_filter_df = df.groupby('class').get_group(x)
        display(x)
        display(age_filter_df)
```
