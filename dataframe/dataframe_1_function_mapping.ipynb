## 1) series.apply(function)

```python
import seaborn as sns

titanic = sns.load_dataset('titanic')
titanic
```

```python
df = titanic.loc[:, ['age', 'fare']]
df['ten'] = 10
df
```

```python
def add_10(n):
    return n + 10

def add_two_obj(a, b):
    return a + b

print(add_10(10))
print(add_two_obj(10, 10))
```

```python
sr1 = df['age'].apply(add_10)   # n=all elements in df['age']
sr1.head()
```

```python
sr2 = df['age'].apply(add_two_obj, b=10)   # a=all elements in df['age'], b=10
sr2.head()
```

```python
sr3 = df['age'].apply(lambda x: add_10(x))   # x=df['age']
sr3.head()
```

```python
sr4 = df['age'].apply(lambda a, b: add_two_obj(a, b), b=10)
sr4.head()
```

## 2) dataframe.applymap(function)

```python
df = titanic.loc[:, ['age', 'fare']]
df
```

```python
def add_10(n):
    return n + 10
```

```python
df1 = df.applymap(add_10)
df1
```

```python
df2 = df.applymap(lambda x: add_10(x))
df2
```

## 3) dataframe.apply(function, axis=0)

```python
df = titanic.loc[:, ['age', 'fare']]
df
```

```python
# input=series, output=series
def missing_value(series):
    return series.isna()
```

```python
df1 = df.apply(missing_value, axis=0)
df1
```

```python
df2 = df.apply(missing_value)   # axis=0 is default
df2
```

```python
df3 = df.apply(lambda x: missing_value(x), axis=0)
df3 
```

```python
df4 = df.apply(lambda x: missing_value(x))   # axis=0 is default
df4 
```

```python
# input=series, output=one value
def min_max(x):
    return (x - x.min()) / (x.max() - x.min())
```

```python
df1 = df.apply(min_max)   # axis=0 is default
df1
```

```python
df1['age'].describe()
```

```python
df1['fare'].describe()
```

## 5) dataframe.pipe(function)

```python
df = titanic.loc[:, ['age', 'fare']]
df
```

```python
# input: df, output: df
def missing_value(x):
    return x.isna()

# input: df, output: series
def missing_count(x):
    return missing_value(x).sum()

# input: df, output: single value
def total_number_missing(x):
    return missing_count(x).sum()
```

```python
df.pipe(missing_value)
```

```python
df.pipe(missing_count)
```

```python
df.pipe(total_number_missing)
```
