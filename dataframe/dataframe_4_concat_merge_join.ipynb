## 1) pd.concat([df1, df2, ...])

```python
import pandas as pd

df1 = pd.DataFrame({'a': ['a0', 'a1', 'a2', 'a3'],
                    'b': ['b0', 'b1', 'b2', 'b3'],
                    'c': ['c0', 'c1', 'c2', 'c3']},
                    index=[0, 1, 2, 3])
 
df2 = pd.DataFrame({'a': ['a2', 'a3', 'a4', 'a5'],
                    'b': ['b2', 'b3', 'b4', 'b5'],
                    'c': ['c2', 'c3', 'c4', 'c5'],
                    'd': ['d2', 'd3', 'd4', 'd5']},
                    index=[2, 3, 4, 5])
```

```python
df1
```

```python
df2
```

```python
result1 = pd.concat([df1, df2])
result1
```

```python
result2 = pd.concat([df1, df2], ignore_index=True)
result2
```

```python
result3 = pd.concat([df1, df2], axis=1)
result3
```

```python
result_3_inner = pd.concat([df1, df2], axis=1, join='inner')
result_3_inner
```

```python
sr1 = pd.Series(['e0', 'e1', 'e2', 'e3'], name='e')
sr2 = pd.Series(['f0', 'f1', 'f2'], name='f', index=[3, 4, 5])
sr3 = pd.Series(['g0', 'g1', 'g2', 'g3'], name='g')
```

```python
sr1
```

```python
sr2
```

```python
sr3
```

```python
result4 = pd.concat([df1, sr1], axis=1)
result4
```

```python
result5 = pd.concat([df2, sr2], axis=1, sort=True)
result5
```

```python
result6 = pd.concat([sr1, sr3], axis=1)
result6
```

```python
result7 = pd.concat([sr1, sr3], axis=0, ignore_index=True)
result7
```

## 2) pd.merge(left=, right=, how=, on=)

```python
import pandas as pd

df1 = pd.read_excel('data/stock price.xlsx', engine='openpyxl')
df2 = pd.read_excel('data/stock valuation.xlsx', engine='openpyxl')
```

```python
df1
```

```python
df2
```

```python
merge_inner = pd.merge(df1, df2)
merge_inner
```

```python
merge_outer = pd.merge(df1, df2, how='outer', on='id')
merge_outer
```

```python
merge_left = pd.merge(df1, df2, how='left', left_on='stock_name', right_on='name')
merge_left
```

```python
merge_right = pd.merge(df1, df2, how='right', left_on='stock_name', right_on='name')
merge_right
```

```python
price = df1[df1['price'] < 50000]
price
```

```python
pd.merge(price, df2)
```

## 3) df1.join(df2, how=)

```python
import pandas as pd

df1 = pd.read_excel('data/stock price.xlsx', engine='openpyxl')
df2 = pd.read_excel('data/stock valuation.xlsx', engine='openpyxl')
```

```python
df1
```

```python
df2
```

```python
df3 = df1.join(df2.set_index('id'), how='left', on='id')
df3
```

```python
df4 = df1.join(df2.set_index('id'), how='inner', on='id')
df4
```
